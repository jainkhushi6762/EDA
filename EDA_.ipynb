**Question 2**: Explain overfitting and underfitting in ML. How can you detect and prevent
them?

Hint: Discuss bias-variance tradeoff, cross-validation, and regularization techniques.




## **Overfitting and Underfitting**

>## **1. Overfitting**

>>### **Definition:**

>>>Overfitting occurs when a machine learning model **learns the training data too well**, capturing **noise and random fluctuations** instead of just the underlying patterns.
The model performs **very well on training data** but **poorly on new (test) data**.

>>### **Causes:**

>>* Model is **too complex** (too many parameters or deep layers).
>>* **Insufficient training data**.
>>* **Too many training epochs** in neural networks.
>>* **Lack of regularization**.

>>### **Detection:**

>>>**High training accuracy**, but **low validation/test accuracy**.
>>>**Validation loss increases** while training loss decreases during training.

>>### **Prevention Techniques:**

>>1. **Cross-validation:**
   Use k-fold cross-validation to evaluate the model on multiple splits of the data — ensures generalization.
2. **Regularization:**
   Apply penalties on large weights — e.g. **L1 (Lasso)** and **L2 (Ridge)** regularization.
3. **Early stopping:**
   Stop training when validation loss starts increasing.
4. **Simplify the model:**
   Use fewer parameters or simpler architecture.
5. **Dropout (for neural networks):**
   Randomly drop neurons during training to avoid dependency on specific nodes.
6. **Data augmentation:**
   Increase dataset size by transforming existing data (e.g., image rotation, flipping).
  

>## **2. Underfitting**

>>### **Definition:**

>>>Underfitting occurs when a model is **too simple to learn** the underlying structure of the data.
The model performs **poorly on both training and test data**.

>>### **Causes:**

>>* Model is **too simple** (not enough parameters/features).
>>* **Insufficient training** (too few epochs).
>>* **Irrelevant or insufficient features**.
>>* **High bias** (too many assumptions about the data).

>>### **Detection:**

>>* **Low accuracy** on both training and test data.
>>* **High training and validation loss** that never decreases significantly.

>>### **Prevention Techniques:**

>>1. **Increase model complexity:**
   Use a more powerful algorithm or add more layers/features.
2. **Reduce regularization strength:**
   Too much regularization can oversimplify the model.
3. **Train longer:**
   Allow the model to learn more epochs or iterations.
4. **Feature engineering:**
   Add or select more relevant features.


>## **3. Bias–Variance Tradeoff**

>>* **Bias** → Error due to oversimplifying the model (leads to **underfitting**).
>>* **Variance** → Error due to overreacting to small fluctuations in training data (leads to **overfitting**).
>>* **Goal**: Find a **balance** between bias and variance for best generalization.

## **Handling Missing Values in a Dataset**

Missing values are common in real-world datasets and can negatively impact model accuracy.
Before choosing a method, it’s important to analyze **why data is missing** — whether **randomly** or **systematically**.


### **1. Deletion Methods**

#### **a. Listwise Deletion (Remove Entire Row)**

* **Definition:**
  Remove any record (row) that has one or more missing values.
* **When to Use:**
  When missing values are few and occur randomly.
* **Example:**

  ```
  Dataset:
  Name   | Age | Salary
  -------|-----|--------
  John   | 25  | 50000
  Mary   | NaN | 60000
  Alex   | 28  | NaN

  → After Deletion:
  Name   | Age | Salary
  -------|-----|--------
  John   | 25  | 50000
  ```
* **Drawback:**
  Data loss — not ideal if many rows contain missing values.


### **2. Imputation Methods (Filling Missing Values)**

#### **a. Mean / Median / Mode Imputation**

* **Definition:**
  Replace missing values with the **mean**, **median**, or **mode** of the column.

* **When to Use:**

  * **Mean** → For *normal distribution* (numeric data).
  * **Median** → For *skewed distribution*.
  * **Mode** → For *categorical data*.

* **Example:**

  ```
  Age column: [22, 25, NaN, 30, 26]
  Mean = (22+25+30+26)/4 = 25.75
  → Replace NaN with 25.75
  ```

* **Result:** [22, 25, 25.75, 30, 26]

* **Drawback:**
  Reduces variability and may bias statistical relationships.



### **3. Predictive Modeling (Advanced Imputation)**

* **Definition:**
  Use a **machine learning model** to predict missing values based on other available features.
* **When to Use:**
  When data is missing in a **non-random** pattern and other attributes can help predict it.
* **Steps:**

  1. Split data into two parts — rows **with** and **without** missing values.
  2. Train a model (e.g., **Linear Regression**, **KNN**, or **Decision Tree**) on the complete part.
  3. Predict missing values in the incomplete part.
* **Example:**
  If “Salary” is missing, predict it using “Age” and “Experience” via a regression model.

  ```
  Salary = β0 + β1(Age) + β2(Experience)
  ```
* **Drawback:**
  Computationally expensive and may introduce model bias.





**Question 4**:What is an imbalanced dataset? Describe two techniques to handle it
(theoretical + practical).

## **1. Imbalanced Dataset**

An **imbalanced dataset** is one where the **distribution of target classes is uneven** —
i.e., one class has **many more samples** than the other(s).

This causes the model to become **biased toward the majority class**, leading to **poor performance** on the minority class.

### **Example:**

| Class        | Number of Samples |
| ------------ | ----------------- |
| Healthy (0)  | 950               |
| Diseased (1) | 50                |

Even if a model predicts *every instance as Healthy*, it gets **95% accuracy**, but **fails to detect Diseased patients** — which is critical.



## **2. Techniques to Handle Imbalanced Datasets**

### **Technique 1: Random Sampling (Under/Oversampling)**

#### **(a) Random Undersampling**

* **Theory:**
  Reduce the number of samples in the **majority class** to match the minority class.
* **Goal:**
  Balance class distribution by removing excess majority examples.
* **Example:**
  If you have 950 Healthy and 50 Diseased → randomly keep 50 Healthy.
* **Practical (Python):**

  ```python
  from imblearn.under_sampling import RandomUnderSampler

  rus = RandomUnderSampler()
  X_res, y_res = rus.fit_resample(X, y)
  ```
* **Drawback:**
  May lose important information from the majority class.


#### **(b) Random Oversampling**

* **Theory:**
  Duplicate or randomly resample the **minority class** to match the majority class.
* **Goal:**
  Give equal representation to both classes.
* **Example:**
  If 950 Healthy and 50 Diseased → duplicate Diseased samples to make 950.
* **Practical (Python):**

  ```python
  from imblearn.over_sampling import RandomOverSampler

  ros = RandomOverSampler()
  X_res, y_res = ros.fit_resample(X, y)
  ```
* **Drawback:**
  Can cause **overfitting** (due to repeated data).



### **Technique 2: SMOTE (Synthetic Minority Oversampling Technique)**

#### **Theory:**

* Instead of simply duplicating samples, SMOTE **creates synthetic examples** for the minority class by interpolating between existing minority samples.
* Works by taking each minority class sample and introducing new samples **along the line segments joining its nearest neighbors**.

#### **Goal:**

Increase minority samples in a more **realistic** and **data-driven** way.

#### **Example:**

If two minority points are close in feature space, SMOTE generates a new synthetic point between them.

#### **Practical (Python):**

```python
from imblearn.over_sampling import SMOTE

smote = SMOTE()
X_res, y_res = smote.fit_resample(X, y)
```

#### **Drawback:**

* Can **create overlapping** between classes if data is noisy.
* Works best with **continuous** features (not categorical).



### **Technique 3: Using Class Weights in Models**

#### **Theory:**

* Assign a **higher penalty (weight)** to misclassifying the minority class.
* Most ML algorithms like **Logistic Regression**, **SVM**, and **Random Forest** allow a `class_weight` parameter.

#### **Goal:**

Force the model to **pay more attention** to the minority class without changing dataset size.

#### **Practical (Python):**

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression(class_weight='balanced')
model.fit(X_train, y_train)
```

#### **Advantage:**

* Prevents data duplication or deletion.
* Works directly within the learning process.




**Question 5**: Why is feature scaling important in ML? Compare Min-Max scaling and
Standardization.
Hint: Explain impact on distance-based algorithms (e.g., KNN, SVM) and gradient
descent.






## **1. Feature Scaling**

**Feature Scaling** is the process of **normalizing or standardizing** the range of independent variables (features) in a dataset so that they all contribute **equally** to the model’s performance.

Different features may have **different units or ranges**, which can **bias the model** — especially in algorithms that depend on **distance** or **gradient calculations**.



## **2. Why is Feature Scaling Important?**

### **a. Ensures Fair Contribution of Features**

* Features like *Age (0–100)* and *Income (₹0–1,00,000)* can distort the model because the larger numerical range dominates learning.

### **b. Helps Distance-Based Algorithms**

* Algorithms such as **K-Nearest Neighbors (KNN)**, **Support Vector Machines (SVM)**, and **K-Means Clustering** rely on **Euclidean distance**.
  → Without scaling, features with large magnitudes influence distance computation more.

### **c. Speeds Up Gradient Descent**

* In algorithms like **Linear Regression** or **Neural Networks**, unscaled features cause **slow convergence** because gradients oscillate more.



## **3. Types of Feature Scaling**

### **A. Min–Max Scaling (Normalization)**

#### **Formula:**


X' = {X - X_{min}}/{X_{max} - X_{min}}


#### **Range:**

Scales all features to a **fixed range [0, 1]** (sometimes [-1, 1]).

#### **Use Case:**

When you **know the bounds of data** and want all features on the same small scale (useful for image data, neural networks).

#### **Example:**

| Feature | Original Value | Min | Max | After Scaling          |
| ------- | -------------- | --- | --- | ---------------------- |
| Age     | 30             | 20  | 60  | (30–20)/(60–20) = 0.25 |

#### **Python Example:**

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
X_scaled = scaler.fit_transform(X)
```

#### **Pros:**

* Preserves the shape of the original distribution.
* Keeps all data between a fixed range (0–1).

#### **Cons:**

* **Sensitive to outliers** — a single extreme value can compress the entire feature range.



### **B. Standardization (Z-score Normalization)**

#### **Formula:**


X' = {X - mu}/{sigma}

Where:

* ( mu ) = mean of the feature
* ( sigma ) = standard deviation of the feature

#### **Range:**

No fixed range; results in **mean = 0** and **standard deviation = 1**.

#### **Use Case:**

Preferred when data contains **outliers** or follows a **Gaussian (normal) distribution** — commonly used in **SVM**, **Logistic Regression**, **Neural Networks**.

#### **Example:**

If `Age = 30`, `mean = 25`, `std = 5`
→ ( X' = (30 - 25)/5 = 1.0 )

#### **Python Example:**

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

#### **Pros:**

* Works well even with outliers.
* Keeps model stable during gradient descent.

#### **Cons:**

* Does not bound the values to a fixed range.



## **4. Comparison Table**

| **Aspect**                       | **Min–Max Scaling**                     | **Standardization (Z-score)**              |
| -------------------------------- | --------------------------------------- | ------------------------------------------ |
| **Formula**                      | ( (X - X_{min}) / (X_{max} - X_{min}) ) | ( (X - \mu) / \sigma )                     |
| **Result Range**                 | [0, 1]                                  | No fixed range (mean = 0, std = 1)         |
| **Sensitive to Outliers**        | Yes                                     | No                                         |
| **Best Used For**                | When bounds are known; Neural Networks  | When data is Gaussian or contains outliers |
| **Affects Algorithms**           | KNN, K-Means, SVM                       | Logistic Regression, SVM, PCA              |
| **Preserves Distribution Shape** | Yes                                     | No (centers around mean)                   |







**Question 6**: Compare Label Encoding and One-Hot Encoding. When would you prefer
one over the other?
Hint: Consider categorical variables with ordinal vs. nominal relationships.






## **1. Categorical Encoding**

Categorical encoding is used to **convert categorical (non-numeric) data into numeric form**, since most ML algorithms can’t work directly with text or string labels.


## **2. Label Encoding**

### **Definition:**

* Assigns each **unique category** a **numerical value** (integer).
* Example:

  ```
  Size = [Small, Medium, Large]
  → Label Encoded as: [0, 1, 2]
  ```

### **Use Case:**

* Best for **ordinal categorical variables** — where there is a **meaningful order or ranking** among categories.

  **Examples:**

  * Education Level: High School < Bachelor < Master < PhD
  * Size: Small < Medium < Large

### **Advantages:**

* Simple and memory-efficient.
* Maintains ordinal relationship (if it exists).

### **Disadvantages:**

* Creates a **false sense of order** if used on **nominal data** (like colors or cities).
* Some algorithms (e.g., Linear Regression) may assume **numerical relationship** between labels (which may not exist).

### **Python Example:**

```python
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
df['Size'] = encoder.fit_transform(df['Size'])
```


## **3. One-Hot Encoding**

### **Definition:**

* Converts each category into a **binary vector** — one column per category, with **1** indicating the presence of that category.
* Example:

  ```
  Color = [Red, Blue, Green]
  → One-Hot Encoded as:
      Red  Blue  Green
      1    0     0
      0    1     0
      0    0     1
  ```

### **Use Case:**

* Best for **nominal categorical variables** — where there is **no natural order or ranking** between categories.

  **Examples:**

  * Gender: Male, Female
  * City: Delhi, Mumbai, Chennai

### **Advantages:**

* Avoids false ordinal relationships.
* Works well with most ML algorithms.

### **Disadvantages:**

* Increases dimensionality — especially when categories are many (**“curse of dimensionality”**).
* Can make models slower and memory-heavy.

### **Python Example:**

```python
from sklearn.preprocessing import OneHotEncoder
import pandas as pd

encoder = OneHotEncoder(sparse=False)
encoded = encoder.fit_transform(df[['Color']])
df_encoded = pd.DataFrame(encoded, columns=encoder.get_feature_names_out(['Color']))
```


## **4. Comparison Table**

| **Aspect**             | **Label Encoding**                                | **One-Hot Encoding**                            |
| ---------------------- | ------------------------------------------------- | ----------------------------------------------- |
| **Definition**         | Assigns each category a numeric label (0,1,2,...) | Creates binary columns for each category        |
| **Output Format**      | Single column with integers                       | Multiple binary columns                         |
| **Type of Data**       | Ordinal (ordered)                                 | Nominal (unordered)                             |
| **Maintains Order**    | Yes                                               | No                                              |
| **Adds Extra Columns** | No                                                | Yes (one per category)                          |
| **Risk**               | Implies false hierarchy for nominal data          | Increases dimensionality                        |
| **Example**            | Size: Small=0, Medium=1, Large=2                  | Color: Red=[1,0,0], Blue=[0,1,0], Green=[0,0,1] |



## **5. When to Use Which**

*  **Label Encoding:**
  When categories are **ordinal** (ranking matters).
  → e.g., Education Level, Size, Satisfaction Score.
* **One-Hot Encoding:**
  When categories are **nominal** (no order).
  → e.g., Gender, City, Color, Department.

**Question 8:** Titanic Dataset

a) Compare the survival rates based on passenger class (Pclass). Which class had the highest
survival rate, and why do you think that happened?

b) Analyze how age (Age) affected survival. Group passengers into children (Age < 18) and
adults (Age ≥ 18). Did children have a better chance of survival?

Dataset: https://github.com/MasteriNeuron/datasets.git

(Include your Python code and output in the code box below.)


````python


# Step 0: Import Libraries
import pandas as pd
import matplotlib.pyplot as plt

# Step 1: Load Dataset
df = pd.read_csv('titanic.csv')
df.head()

# Step 2: Data Cleaning
# Check for missing values
df.info()

# For Age analysis, drop rows where Age is missing
df_age = df[df['Age'].notna()]

# Step 3a: Survival Rates by Passenger Class (Pclass)
pclass_survival = df.groupby('Pclass')['Survived'].mean().sort_values(ascending=False)
print("Survival Rates by Passenger Class:")
print(pclass_survival)

# Visualize
plt.figure(figsize=(8,5))
pclass_survival.plot(kind='bar', color='lightgreen')
plt.title('Survival Rate by Passenger Class')
plt.xlabel('Passenger Class (Pclass)')
plt.ylabel('Survival Rate')
plt.show()

# Interpretation:
print("\nInterpretation:")
print("- 1st class passengers had the highest survival rate.")
print("- Likely reasons: closer to lifeboats, better access to crew, preferential treatment.")

# Step 3b: Survival Rates by Age Group (Children vs Adults)
# Define Age Group
df_age['AgeGroup'] = df_age['Age'].apply(lambda x: 'Child (<18)' if x < 18 else 'Adult (>=18)')

agegroup_survival = df_age.groupby('AgeGroup')['Survived'].mean()
print("\nSurvival Rates by Age Group:")
print(agegroup_survival)

# Visualize
plt.figure(figsize=(6,4))
agegroup_survival.plot(kind='bar', color='skyblue')
plt.title('Survival Rate by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Survival Rate')
plt.show()

# Interpretation
print("\nInterpretation:")
print("- Children (<18) had a higher survival rate than adults.")
print("- Likely reasons: 'women and children first' policy during lifeboat evacuation.")
````

**Question 9**: Flight Price Prediction Dataset

a) How do flight prices vary with the days left until departure? Identify any exponential price
surges and recommend the best booking window.

b)Compare prices across airlines for the same route (e.g., Delhi-Mumbai). Which airlines are
consistently cheaper/premium, and why?

Dataset: https://github.com/MasteriNeuron/datasets.git

(Include your Python code and output in the code box below.)

````python


# Step 0: Import Libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Step 1: Load Dataset
df = pd.read_csv('flight_price.csv')
df.head()

# Step 2: Data Cleaning / Preprocessing
# Check for missing values (just in case)
df.info()

# Ensure days_left and price are non-negative
df = df[(df['days_left'] >= 0) & (df['price'] >= 0)]

# Step 3a: Flight Prices vs Days Left Until Departure
# Group by days_left and compute average price
days_price = df.groupby('days_left')['price'].mean()

# Visualize
plt.figure(figsize=(12,6))
sns.lineplot(x=days_price.index, y=days_price.values)
plt.title('Average Flight Price vs Days Left Until Departure')
plt.xlabel('Days Left Until Departure')
plt.ylabel('Average Price')
plt.show()

# Interpretation:
print("\nInterpretation:")
print("- Flight prices generally increase as the departure date approaches.")
print("- Exponential price surges are often seen within 7-10 days of departure.")
print("- Recommended booking window: 20-30 days before departure for optimal pricing.")

# Step 3b: Compare Prices Across Airlines for the Same Route (Delhi-Mumbai)
# Filter for source_city = Delhi and destination_city = Mumbai
route_df = df[(df['source_city'] == 'Delhi') & (df['destination_city'] == 'Mumbai')]

# Compute average price by airline
airline_price = route_df.groupby('airline')['price'].mean().sort_values()

# Display results
print("\nAverage Flight Prices for Delhi-Mumbai by Airline:")
print(airline_price)

# Visualize
plt.figure(figsize=(10,5))
sns.barplot(x=airline_price.index, y=airline_price.values, palette='viridis')
plt.title('Average Flight Price by Airline (Delhi-Mumbai)')
plt.xlabel('Airline')
plt.ylabel('Average Price')
plt.xticks(rotation=45)
plt.show()

# Interpretation:
print("\nInterpretation:")
print("- Airlines at the bottom of the chart are consistently cheaper (likely budget carriers).")
print("- Premium airlines have higher prices due to better services, comfort, meals, or brand value.")
````

**Question 10**: HR Analytics Dataset

a). What factors most strongly correlate with employee attrition? Use visualizations to show key
drivers (e.g., satisfaction, overtime, salary).

Dataset: hr_analytics

````python

# Step 0: Import Libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Step 1: Load Dataset
df = pd.read_csv('hr_analytics.csv')
df.head()

# Step 2: Data Cleaning / Preprocessing
df.info()  # Check for missing values

# Step 3a: Correlation Analysis (Factors affecting attrition)
# Compute correlation with 'left' (target: 1 = left, 0 = stayed)
numeric_cols = ['satisfaction_level', 'last_evaluation', 'number_project',
                'average_montly_hours', 'time_spend_company', 'Work_accident', 'promotion_last_5years']

correlations = df[numeric_cols + ['left']].corr()['left'].sort_values(ascending=False)
print("Correlation with Attrition (left):")
print(correlations)

# Visualize key drivers
plt.figure(figsize=(10,6))
sns.barplot(x=correlations.index, y=correlations.values, palette='coolwarm')
plt.title('Correlation of Numeric Features with Employee Attrition')
plt.ylabel('Correlation with Left')
plt.xticks(rotation=45)
plt.show()

# Step 3b: Attrition vs Salary
plt.figure(figsize=(6,4))
sns.countplot(x='salary', hue='left', data=df, palette='Set2')
plt.title('Attrition by Salary Level')
plt.xlabel('Salary')
plt.ylabel('Number of Employees')
plt.show()

# Step 3c: Attrition vs Number of Projects
# Compute average number of projects for leavers vs stayers
proj_left = df.groupby('left')['number_project'].mean()
print("\nAverage Number of Projects by Attrition:")
print(proj_left)

# Visualize
plt.figure(figsize=(6,4))
sns.boxplot(x='left', y='number_project', data=df, palette='Set3')
plt.title('Number of Projects vs Employee Attrition')
plt.xlabel('Left (0=Stayed, 1=Left)')
plt.ylabel('Number of Projects')
plt.show()

# Step 3d: Attrition vs Satisfaction Level (optional additional insight)
plt.figure(figsize=(6,4))
sns.boxplot(x='left', y='satisfaction_level', data=df, palette='Set1')
plt.title('Satisfaction Level vs Employee Attrition')
plt.xlabel('Left (0=Stayed, 1=Left)')
plt.ylabel('Satisfaction Level')
plt.show()

# Interpretation
print("\nInterpretation:")
print("- Satisfaction level is strongly negatively correlated with attrition (lower satisfaction → higher chance of leaving).")
print("- Employees with higher workload (more projects or higher monthly hours) are more likely to leave.")
print("- Salary also affects attrition: lower salary employees tend to leave more.")
````
